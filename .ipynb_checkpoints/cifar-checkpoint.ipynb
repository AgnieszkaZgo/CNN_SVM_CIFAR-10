```python
import pandas as pd
from keras.datasets import cifar10
import numpy as np
import matplotlib.pyplot as plt
from itertools import product
from keras.models import Sequential 
from keras.layers import Input, Dense, Flatten
from keras.optimizers import Adam
from keras.utils import to_categorical
from sklearn import svm
from keras.applications.resnet50 import ResNet50
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
```

```python
(x_train, y_train), (x_test, y_test) = cifar10.load_data()
# y_train = to_categorical(y_train)
# y_test = to_categorical(y_test)
```

```python
x_train = x_train[:5000]
y_train = y_train[:5000]
x_test = x_test[:1000]
y_test = y_test[:1000]
```

```python
data = pd.DataFrame()
data['images']=pd.Series(list(x_train))
data['labels']=pd.Series(y_train.flatten())
to_print=[None]*10
for i in range(10):
    to_print[i] = data[data.labels==i].sample(10).images.reset_index(drop=True)
```

```python
_, axs = plt.subplots(10,10, figsize=(12,12))

for i,j in product(range(10),repeat=2):
    axs[i,j].imshow(to_print[i][j])
    axs[i,j].axis('off')
```

```python
shallow = Sequential()
shallow.add(Flatten(input_shape=(32,32,3)))
shallow.add(Dense(1000, activation='relu'))
shallow.add(Dense(10, activation='softmax'))
shallow.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['categorical_accuracy'])
```

```python
shallow.fit(x_train, y_train, epochs=10)
```
