```python
import pandas as pd
from keras.datasets import cifar10
import numpy as np
import matplotlib.pyplot as plt
from itertools import product
from keras.models import Sequential 
from keras.layers import Input, Dense, Flatten, GlobalAveragePooling2D
from keras.optimizers import Adam
from keras.utils import to_categorical
from sklearn import svm
from keras.applications.resnet50 import ResNet50
from keras.applications.inception_v3 import InceptionV3
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from PIL import Image
from sklearn.model_selection import GridSearchCV
```

```python
(x_train, y_train), (x_test, y_test) = cifar10.load_data()
# y_train = to_categorical(y_train)
# y_test = to_categorical(y_test)
```

```python
x_train = x_train[:1000]
y_train = y_train[:1000]
x_test = x_test[:200]
y_test = y_test[:200]
```

```python
data = pd.DataFrame()
data['images']=pd.Series(list(x_train))
data['labels']=pd.Series(y_train.flatten())
to_print=[None]*10
for i in range(10):
    to_print[i] = data[data.labels==i].sample(10).images.reset_index(drop=True)
```

```python
_, axs = plt.subplots(10,10, figsize=(12,12))

for i,j in product(range(10),repeat=2):
    axs[i,j].imshow(to_print[i][j])
    axs[i,j].axis('off')
```

```python
shallow = Sequential()
shallow.add(Flatten(input_shape=(32,32,3)))
shallow.add(Dense(1000, activation='relu'))
shallow.add(Dense(10, activation='softmax'))
shallow.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['categorical_accuracy'])
shallow.fit(x_train, y_train, epochs=10)
```

```python
model_res = ResNet50(weights='imagenet', include_top=False, input_shape=(32,32,3), classes=10)
```

```python
codes_train = model_res.predict(x_train).reshape(x_train.shape[0],-1)
codes_test = model_res.predict(x_test).reshape(x_test.shape[0],-1)
```

```python
scaler = StandardScaler()
scaler.fit(codes_train)
codes_train = scaler.transform(codes_train)
svc = svm.LinearSVC(C=10)
svc.fit(codes_train, y_train.flatten())
```

```python
codes_test = scaler.transform(codes_test)
pred = svc.predict(codes_test)
accuracy_score(pred, y_test.flatten())
```

```python
from tsne import bh_sne

Y = bh_sne(codes_test[:2000].astype('float64'))
print('+')
min_1 = Y[:,0].min()
max_1 = Y[:,0].max()
min_2 = Y[:,1].min()
max_2 = Y[:,1].max()
Yn = Y[:]
Yn[:,0] = (Y[:,0] - min_1) / (max_1 - min_1) 
Yn[:,1] = (Y[:,1] - min_2) / (max_2 - min_2)
print('+')
plt.scatter(Yn[:, 1], Yn[:, 0], c=y_test.flatten()[:2000], cmap=plt.cm.get_cmap("jet", 10), s=4, edgecolors='none')
plt.colorbar(ticks=range(10))
```
